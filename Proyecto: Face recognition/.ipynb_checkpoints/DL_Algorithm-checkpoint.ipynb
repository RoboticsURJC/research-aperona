# Reconocimiento de caras usando un algoritmo de Deep Learning

## Paso 0: Documentación:

En este proyecto, implementaremos un algoritmo de reconocimiento de caras basado en **redes neuronales**, comparando sus prestaciones con distintos dispositivos hardware de captura de imágenes.
Para ello, emplearemos las siguientes **librerías** de python:

* Para tratamiento de imágenes:

    - PIL
    - OpenCV 
    - matplotlib


* Para creación de redes neuronales:

    - torch y facenet_pytorch
    - numpy
    
* Para seguimiento del tiempo de ejecucuión:

    - time


```python
#-- Tratamiento de imágenes
from PIL import Image
import cv2
import matplotlib.pyplot as plt

#-- Redes neuronales
import torch
from facenet_pytorch import MTCNN
import numpy as np

#-- Tomatiempo
import time
```

## Paso 2: Reconocimiento de caras en una WebCam

```python
"""
	En este caso, 0 quiere decir que queremos acceder
	a la cámara 0. Si hay más cámaras, puedes ir probando
	con 1, 2, 3...
"""


inicio = time.time()

cap = cv2.VideoCapture(0)

leido, frame = cap.read()

if leido == True:
    #--cv2.imwrite("foto.png", frame)
    print("Foto tomada correctamente")
else:
    print("Error al acceder a la cámara")

"""
    Finalmente liberamos o soltamos la cámara
"""

try:
    # Detección de bounding box y landmarks
    # ==============================================================================
    boxes, probs, landmarks = mtcnn.detect(frame, landmarks=True)

    # Representación con matplotlib
    # ==============================================================================
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.imshow(frame)

    for box, landmark in zip(boxes, landmarks):
        ax.scatter(landmark[:, 0], landmark[:, 1], s=8, c= 'red')
        rect = plt.Rectangle(
                    xy     = (box[0], box[1]),
                    width  = box[2] - box[0],
                    height = box[3] - box[1],
                    fill   = False,
                    color  = 'red'
               )
        ax.add_patch(rect)

    fin = time.time()

    ex_time = fin - inicio

    print(f"Tiempo de ejecución: {ex_time} segundos")

    ax.axis('off');
    
    cap.close()

except AttributeError:
    print("ERROR: Ejecute de nuevo el código")
```
